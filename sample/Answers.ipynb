```python
df['Start Time'] = pd.to_datetime(df['Start Time'], format='%m/%d/%Y %H:%M')
df['Start Time']
df['End Time'] = pd.to_datetime(df['End Time'], format='%m/%d/%Y %H:%M')
df
```

```python
df['Duration_hour']= df.loc[:,'End Time'].dt.hour - df.loc[:,'Start Time'].dt.hour
df['Duration_min']= df.loc[:,'End Time'].dt.minute - df.loc[:,'Start Time'].dt.minute
lst=[]
for i in df['Duration_min']:
    if(i<0):
        i=60+i
        lst.append(i)
    else:
        lst.append(i)
df['Duration_min']= lst
df['Duration(s)']= df.loc[:,'Duration_hour']*3600 + df.loc[:,'Duration_min']*60
df['Duration(s)']
#(df['End Time'] - df['Start Time']) / pd.Timedelta(seconds=1)
```

```python
df['WEEKDAY'] = pd.to_datetime(df['Start Time']).dt.dayofweek
lst=[]
for i in df['WEEKDAY']:
    if(i==5 or i==6):
        lst.append("Yes")
    else:
        lst.append("No")
df['is_weekend']=lst
df['is_weekend']
```

```python
df['hour_of_the_day']= df['Start Time'].dt.hour
df['hour_of_the_day']
```

```python
from math import *
df['Start Latitude_rad']= df.loc[:,'Start Latitude']*(180/pi)
df['End Latitude_rad']= df.loc[:,'End Latitude']*(180/pi)
df['Start Longitude_rad']= df.loc[:,'Start Longitude']*(180/pi)
df['End Longitude_rad']= df.loc[:,'End Longitude']*(180/pi)
df['dlat']=df['End Latitude_rad']-df['Start Latitude_rad']
df['dlon']=df['End Longitude_rad']-df['Start Longitude_rad']

```

```python
a=[]
c=[]
dist=[]
for i in range(len(df)):
    a.append(sin(df.loc[i,'dlat'] / 2)**2 + cos(df.loc[i,'Start Latitude_rad']) * cos(df.loc[i,'End Latitude_rad']) * sin(df.loc[i,'dlon'] / 2)**2)
    c.append(2 * asin(sqrt(a[i])))
    dist.append(c[i]*6371)
df['distance(km)']= dist
```

```python
df = df.drop(['Start Latitude_rad','End Latitude_rad','Start Longitude_rad', 'End Longitude_rad','dlat', 'dlon','WEEKDAY','Duration_hour','Duration_min'], axis=1)
```
